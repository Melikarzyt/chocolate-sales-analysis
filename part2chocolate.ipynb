```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('chocolate_preprocessed.csv')
```

```python
max = df['Rating'].max()
max
```

```python
df['Rating'] = df['Rating'] * 20
df.head()
```

```python
df.info()
```

```python
df['price(100g)'] = df['Cocoa Percent'] * df['Rating'] * 25
df.head()
```

```python
print(df['Cocoa Percent'].dtype)
```

```python
dark_chocolated = df[df['Cocoa Percent'] > 70].reset_index(drop=True)
dark_chocolated
```

```python
dark_chocolated['has_Trinitario'] = dark_chocolated['Bean Type'].fillna(' ').str.contains('Trinitario')
dark_chocolated
```

```python
dark_chocolated.loc[dark_chocolated['has_Trinitario'] == True, 'price(100g)'] *= 1.1
dark_chocolated[dark_chocolated['has_Trinitario']]
```

```python
pricesum = dark_chocolated.loc[dark_chocolated['has_Trinitario'] == True, 'price(100g)'].sum()
print(pricesum)
```

```python
df.to_csv('chocolate_price.csv', index=False)
```

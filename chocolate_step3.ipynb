```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('chocolate_price.csv')
df.head()
```

```python
df = df[df['Cocoa Percent'] <= 70].reset_index(drop=True)
df
```

```python
companies = df.groupby(['Company', 'Review Date'])['Rating'].mean().unstack()
```

```python
companies.columns.name = None
companies.index.name = None
```

```python
companies
```

```python
mean_ratings = df[df['Review Date'] >= 2012].groupby('Company')['Rating'].mean()
mean_ratings
```

```python
best_ratings = mean_ratings.nlargest(10)
```

```python
print(best_ratings)
```

```python
chocolates_to_sell = df[df['Company'].isin(best_ratings.index) & (df['Review Date'] >= 2012)].reset_index(drop=True)
```

```python
print(chocolates_to_sell)
```

```python
pricesum = chocolates_to_sell['price(100g)'].sum()
print(pricesum)
```
